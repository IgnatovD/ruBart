```
import pandas as pd
import numpy as np
import tokenizers
import torch
from pathlib import Path

from tokenizers.implementations import ByteLevelBPETokenizer
from tokenizers.processors import BertProcessing
```

### Сreate training files

```
for i in range(1, 13):
  path = '../data/txt'
  data_dir = f'{path}/ruwiki_txt_{i}.tar.gz'
  !tar -xf {data_dir}
```

```
path = Path('data')
train_file = [path.as_posix() for path in path.rglob('*.txt')]
```

```
bpe = tokenizers.ByteLevelBPETokenizer(
        vocab_file=None,
        lowercase=True,
        trim_offsets=True
)
```

* Show_progress is not working!
* Full training takes about 45 minutes.
* Wait for processing to complete in the cell.

```
bpe.train(
    files=train_file,
    vocab_size=30000,
    min_frequency=3,
    show_progress=False,
    special_tokens = ["<s>", "<pad>", "</s>", "<unk>", "<mask>"]
)
```

```
bpe.save('.', 'bart')
```

### Test tokenizer

```
tokenizer = ByteLevelBPETokenizer('bart-vocab.json', 'bart-merges.txt')

tokenizer._tokenizer.post_processor = BertProcessing(
    ("</s>", tokenizer.token_to_id("</s>")),
    ("<s>", tokenizer.token_to_id("<s>")),
)
tokenizer.enable_truncation(max_length=512)
```

```
tokenizer.encode('Сегодня тепло', 'крбрфтсп').ids
```
